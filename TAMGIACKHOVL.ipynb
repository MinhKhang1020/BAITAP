```python
import math
xA = int(input('Nhap hoanh do diem A: '))
yA = int(input('Nhap tung do diem A: '))
xB = int(input('Nhap hoanh do diem B: '))
yB = int(input('Nhap tung do diem B: '))
xC = int(input('Nhap hoanh do diem C: '))
yC = int(input('Nhap tung do diem C: '))

#Xet 3 diem A B C khong thang hang

AB = math.sqrt((xB-xA)**2 + (yB-yA)**2)
AC = math.sqrt((xC-xA)**2 + (yC-yA)**2)
BC = math.sqrt((xC-xB)**2 + (yC-yB)**2)
Tamgiac = AB + AC > BC and AB + BC > AC and AC + BC > AB
print('AB, AC, BC la 3 canh cua tam giac', Tamgiac)

cosA = (AC**2 + BC**2 - AB**2)/(2*AC*BC)
cosB = (AB**2 + BC**2 - AC**2)/(2*AB*BC)
cosC = (AB**2 + AC**2 - BC**2)/(2*AB*AC)
A = math.degrees(math.acos(cosA))
B = math.degrees(math.acos(cosB))
C = math.degrees(math.acos(cosC))

#Tinh do dai duong cao tu cac dinh

DuongcaoAH = AH = math.sin(B)*AB
DuongcaoBJ = BJ = math.sin(C)*BC
DuongcaoCK = CK = math.sin(A)*AC

print('Duong cao tu dinh A: ',AH)
print('Duong cao tu dinh B: ',BJ)
print('Duong cao tu dinh C: ',CK)

#Tinh toa do trong tam tam giac ABC
xM = (xA+xB+xC)/3
yM = (yA+yB+yC)/3
print('Toa do trong tam M cua tam giac ABC la: ',(xM, yM))

#Tinh toa do truc tam tam giac ABC
#vector AB, AC, BC
xAB = xB-xA
yAB = yB-yA
xAC = xC-xA
yAC = yC-yA
xBC = xC-xB
yBC = yC-yB
xO = 'Hoanh do truc tam O'
yO = 'Tung do truc tam O'

D1 = xA*xBC + yBC*yA
D2 = xAC*xB + yAC*yB
yO = (D1*xAC-D2*xBC)/(yBC*xAC-yAC*xBC)
xO = (D1-yBC*yO)/xBC
print('Toa do truc tam tam giac ABC la: ',(xO, yO))
import turtle
n = turtle.Turtle()
n.hideturtle()
n.penup()
n.goto(xA, yA)
n.write('A', align='right')
n.pendown()
n.goto(xB, yB)
n.write('B')
n.goto(xC, yC)
n.write('C')
n.goto(xA, yA)
n.penup()
n.pensize(3)
n.goto(xM, yM)
n.pendown()
n.pencolor('red')
n.goto(xM, yM)
n.write('M')
n.penup()
n.goto(xO, yO)
n.pendown()
n.pencolor('blue')
n.goto(xO, yO)
n.write('O')
turtle.done()
```
